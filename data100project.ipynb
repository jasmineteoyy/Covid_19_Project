# Cleaning Data and Making Everything Consistent
### Gregorian Time to Timestamp Consistent Format

```python
rand_timestamp = pd.Timestamp(year = 2011,  month = 11, day = 21, 
                  hour = 10, second = 49, tz = 'US/Central') 

list_of_gregorian_columns = ['>50 gatherings', '>500 gatherings','stay at home', 'public schools','restaurant dine-in', 'entertainment/gym', 'federal guidelines', 'foreign travel ban']
index_of_column = 0 
for col in list_of_gregorian_columns:
    index_of_column = 0 
    for data in county_data[col]:
        if data == None or math.isnan(data):
            index_of_column += 1
            continue
        else:
            county_data.loc[index_of_column, col] = rand_timestamp.fromordinal(ordinal=int(data))
            index_of_column += 1
```

```python
county_data 
```

### Combine county_data with more features from states table

```python
county_and_states = county_data.merge(states, right_on='Province_State', left_on='State', how='inner').drop(columns=['Province_State', 'Country_Region', 'Last_Update', 'Deaths', 'UID', 'ISO3'])
county_and_states
```

### Get rid of columns that won't be features and columns with at least 20% nans

```python
county_and_states = county_and_states.drop(columns=['countyFIPS', 'STATEFP', 'COUNTYFP', 'CountyName', 'StateName', 'federal guidelines', 'foreign travel ban'])
county_and_states = county_and_states.dropna(thresh=len(county_data) - len(county_data)/5, axis=1) #don't want features that cause us to drop many cols
county_and_states #thresh = len - number of nans at most
```

### One Hot Encode Categorical Features in the Table

```python
county_and_states
```

```python
county_and_states = county_and_states.dropna()
def one_hot_encoding(df, column_name):
    one_hot_encoded_column = pd.get_dummies(df[column_name].astype(str),prefix=column_name, dummy_na = False)
    one_hot_encoded = pd.concat([df,one_hot_encoded_column], axis=1).drop(columns=[column_name])
    return one_hot_encoded
list_of_cols_to_ohe = ['CensusDivisionName', 'CensusRegionName','>50 gatherings', '>500 gatherings', 'public schools','restaurant dine-in', 'entertainment/gym']
for col in list_of_cols_to_ohe:
    county_and_states = one_hot_encoding(county_and_states, col)
county_and_states
```

# STANDARDIZING DATAFRAME COLUMNS

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
county_and_states[county_and_states.columns[1:]] = scaler.fit_transform(county_and_states[county_and_states.columns[1:]])
```

```python
county_and_states
```

### Merging Deaths to the County+States Subset Table

```python
deaths_by_state = deaths.reset_index()[['Province_State', deaths.columns[len(deaths.columns)-1]]]

county_states_deaths = deaths_by_state.merge(right=county_and_states, right_on='State', left_on='Province_State', how='inner').drop(columns=['State']).rename(columns={'Province_State':'State'})
county_states_deaths = county_states_deaths.rename(columns={county_states_deaths.columns[1] : 'Deaths - 4/18/2020'})
county_states_deaths
```

```python
# Import the LinearRegression model from sklearn
from sklearn.linear_model import LogisticRegression, Lasso, Ridge
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
def rmse_score(model, X, y):
    return np.sqrt(np.mean((y - model.predict(X)) ** 2))

our_chosen_features_tbl = county_states_deaths[['Deaths - 4/18/2020', 'Smokers_Percentage', 'DiabetesPercentage', 'HeartDiseaseMortality', 'StrokeMortality', '#EligibleforMedicare2018', 'MedianAge2010', 'PopulationEstimate2018']]
our_chosen_features_tbl = our_chosen_features_tbl.dropna()

model1 = Lasso(alpha=0.5)
X_train, X_test, y_train, y_test = train_test_split(our_chosen_features_tbl[our_chosen_features_tbl.columns[1:]], pd.Series(our_chosen_features_tbl[our_chosen_features_tbl.columns[0]]), test_size=0.2, random_state = 85)
model1.fit(X_train, y_train)

cv_error = np.mean(cross_val_score(model1, X_train, y_train, scoring=rmse_score, cv=5))
cv_error
```

# TESTING MULTIPLE RANDOM SUBSETS

```python
def rand_features():
    county_subset = county_states_deaths[['Deaths - 4/18/2020']+ list(random.sample(cols, random.sample(range(6,10), 1)[0]))]
    county_subset = county_subset.dropna()

    

    model1 = Lasso(alpha=0.5)
    X_train, X_test, y_train, y_test = train_test_split(county_subset[county_subset.columns[1:]], pd.Series(county_subset[county_subset.columns[0]]), test_size=0.2, random_state=85)
    model1.fit(X_train, y_train)

    cv_error = np.mean(cross_val_score(model1, X_train, y_train, scoring=rmse_score, cv=5))
    county_cols = list(county_subset.columns)
    county_cols.pop(0)
    return county_cols, cv_error

min_features = []
min_error = float('inf')
for i in range(0, 1000):
    features, cv_error = rand_features()
    if (min_error > cv_error):
        min_error = cv_error
        min_features = features

print(min_features, min_error)
```

# BEST ALPHA 


```python
alpha_arr = np.linspace(0.001, 0.5, 100)
cv_errors = []
X_train, X_test, y_train, y_test = train_test_split(county_states_deaths[['People_Hospitalized', 'CensusDivisionName_East North Central', '>50 gatherings_2020-03-30 00:00:00', 'PopFmle65-742010', 'entertainment/gym_2020-03-20 00:00:00', 'PopMale5-92010', 'restaurant dine-in_2020-03-19 00:00:00', 'Incident_Rate', 'PopMale<52010']], pd.Series(county_states_deaths.iloc[:, 1]), test_size=0.2, random_state=85)



for alpha in alpha_arr:
    
    #model.fit(X_train, Y_train)
    model1 = Lasso(alpha=alpha)

    # compute the cross validation error
    cv_error = np.mean(cross_val_score(model1, X_train, y_train, scoring=rmse_score, cv=5))
    
    cv_errors.append(cv_error)
    
best_alpha_lasso = alpha_arr[np.argmin(cv_errors)]

print(f"The best alpha value is {best_alpha_lasso}")
print(f"Cross validation error for the best alpha value is {cv_errors[np.argmin(cv_errors)]}")
plt.scatter(alpha_arr, cv_errors)
plt.show()
best_alpha_lasso
```

# Final Model + Test Error

```python
X_train, X_test, y_train, y_test = train_test_split(county_states_deaths[['People_Hospitalized', 'CensusDivisionName_East North Central', '>50 gatherings_2020-03-30 00:00:00', 'PopFmle65-742010', 'entertainment/gym_2020-03-20 00:00:00', 'PopMale5-92010', 'restaurant dine-in_2020-03-19 00:00:00', 'Incident_Rate', 'PopMale<52010']], pd.Series(county_states_deaths.iloc[:, 1]), test_size=0.2, random_state=85)
model1 = Lasso(alpha=best_alpha_lasso)
model1.fit(X_train, y_train)

rmse_score(model1, X_test, y_test)
```
